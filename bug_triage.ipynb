```python
# ============================================================
# CELL 2 вЂ” T4-hardened Llama-3.2-1B fine-tuning script
# Run only AFTER restarting runtime following Cell 1.
# Stack: unsloth 2026.6.x | trl 0.21 | torch 2.6+cu118
# ============================================================

# в”Ђв”Ђ 0. Environment guards (must precede ALL torch/unsloth imports) в”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђ

import os, sys

os.environ["ACCELERATE_DISABLE_BF16"] = "1"   # kill BF16 at accelerate level
os.environ["TOKENIZERS_PARALLELISM"]  = "false"  # suppress HF fork deadlock warning

import torch

# Hard stop if no GPU вЂ” catches "forgot to set runtime type" before wasting time
assert torch.cuda.is_available(), (
    "вќЊ No CUDA device. Go to: Runtime в†’ Change runtime type в†’ T4 GPU"
)

gpu_name   = torch.cuda.get_device_name(0)
gpu_mem_gb = torch.cuda.get_device_properties(0).total_memory / 1e9
print(f"вњ… GPU : {gpu_name}  ({gpu_mem_gb:.1f} GB)")

# T4 precision flags:
#   allow_tf32 = True  в†’ safe on T4, free throughput gain for matmul
#   benchmark  = False в†’ deterministic kernel selection; avoids mid-epoch OOM
#                        from cudnn switching to a larger kernel after step 1
torch.backends.cuda.matmul.allow_tf32 = True
torch.backends.cudnn.allow_tf32       = True
torch.backends.cudnn.benchmark        = False
torch.backends.cudnn.deterministic    = True

# в”Ђв”Ђ 1. bitsandbytes smoke test в”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђ
# Probe bnb before loading any model weights. A broken bnb install surfaces here
# as a clean error rather than a cryptic CUDA segfault mid-training.

try:
    import bitsandbytes as bnb
    # Actually exercise the CUDA path, not just the Python import
    _ = bnb.optim.Adam8bit
    print(f"вњ… bitsandbytes {bnb.__version__} вЂ” CUDA path functional")
except Exception as e:
    sys.exit(
        f"вќЊ bitsandbytes failed: {e}\n"
        "   Re-run Cell 1 and restart runtime."
    )

# в”Ђв”Ђ 2. Core imports & version audit в”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђ

import transformers
from unsloth import FastLanguageModel
from trl      import SFTTrainer, SFTConfig
from datasets import load_dataset
import peft, trl, unsloth, accelerate

# Print the full version matrix вЂ” paste this into any future bug report
print("\nв”Ђв”Ђ Version Matrix в”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђ")
for name, mod in [
    ("torch",         torch),
    ("transformers",  transformers),
    ("trl",           trl),
    ("unsloth",       unsloth),
    ("bitsandbytes",  bnb),
    ("peft",          peft),
    ("accelerate",    accelerate),
]:
    print(f"   {name:<16}: {mod.__version__}")
print("в”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђ\n")

# в”Ђв”Ђ 3. Tuneable constants в”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђ

MODEL_NAME      = "unsloth/Llama-3.2-1B-bnb-4bit"
MAX_SEQ_LENGTH  = 512
DTYPE           = torch.float16   # T4 mandatory: no BF16
LOAD_IN_4BIT    = True

LORA_R          = 16
LORA_ALPHA      = 32              # standard: 2Г— r
LORA_DROPOUT    = 0.0             # Unsloth recommends 0 for 4-bit
TARGET_MODULES  = [
    "q_proj", "k_proj", "v_proj", "o_proj",
    "gate_proj", "up_proj", "down_proj",
]

DATA_FILE       = "training_data.jsonl"
OUTPUT_DIR      = "./outputs"
ADAPTER_DIR     = "./lora-adapter"

BATCH_SIZE      = 2               # safe headroom on 15 GB T4 with 1B 4-bit model
GRAD_ACC_STEPS  = 8               # effective batch size = 16
LEARNING_RATE   = 2e-4
NUM_EPOCHS      = 3
WARMUP_STEPS    = 10
LOG_STEPS       = 5

# в”Ђв”Ђ 4. Load base model в”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђ

print("вЏі Loading base model weights...")

model, tokenizer = FastLanguageModel.from_pretrained(
    model_name     = MODEL_NAME,
    max_seq_length = MAX_SEQ_LENGTH,
    dtype          = DTYPE,
    load_in_4bit   = LOAD_IN_4BIT,
)

print("вњ… Base model loaded.")

# в”Ђв”Ђ 5. Tokenizer hardening в”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђ
# Unsloth may set padding_side="left" for inference speed.
# Causal LM training requires "right" padding so loss labels align with tokens.

tokenizer.padding_side = "right"

if tokenizer.pad_token is None:
    # Llama-3 has a dedicated pad token; this branch handles unusual checkpoints.
    tokenizer.pad_token    = tokenizer.eos_token
    tokenizer.pad_token_id = tokenizer.eos_token_id
    print("вљ пёЏ  pad_token was None в†’ aliased to eos_token (safe for causal LM)")

print(f"вњ… Tokenizer: pad='{tokenizer.pad_token}', "
      f"padding_side='{tokenizer.padding_side}'")

# в”Ђв”Ђ 6. Apply LoRA в”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђ

print("\nвЏі Attaching LoRA adapter...")

model = FastLanguageModel.get_peft_model(
    model,
    r              = LORA_R,
    lora_alpha     = LORA_ALPHA,
    lora_dropout   = LORA_DROPOUT,
    target_modules = TARGET_MODULES,
    bias           = "none",
    # "unsloth" = Unsloth's memory-efficient gradient checkpointing variant.
    # Saves ~30% VRAM vs standard HF checkpointing on T4.
    use_gradient_checkpointing = "unsloth",
    random_state   = 42,
    use_rslora     = False,   # rank-stabilised LoRA: off вЂ” fewer variables
    loftq_config   = None,    # no LoftQ: keeps 4-bit quantisation simple
)

trainable = sum(p.numel() for p in model.parameters() if p.requires_grad)
total     = sum(p.numel() for p in model.parameters())
print(f"вњ… LoRA attached: {trainable:,} / {total:,} trainable "
      f"({100 * trainable / total:.3f}%)")

# в”Ђв”Ђ 7. Dataset в”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђ

print("\nвЏі Loading and formatting dataset...")

dataset = load_dataset("json", data_files=DATA_FILE, split="train")

EOS = tokenizer.eos_token  # "<|end_of_text|>" for Llama-3

def format_example(example):
    """
    Produces a single 'text' field per record.

    The EOS token at the end is not cosmetic вЂ” it teaches the model
    where a complete response ends. Without it the model never learns
    to stop generating during inference.
    """
    return {
        "text": (
            f"Instruction: Classify this GitHub issue.\n"
            f"Input: {example['input']}\n"
            f"Output: {example['output']}"
            f"{EOS}"
        )
    }

dataset = dataset.map(
    format_example,
    remove_columns = dataset.column_names,  # discard original cols; keep only 'text'
    num_proc       = 1,    # 1 = main process only; avoids Colab fork() deadlocks
    desc           = "Formatting",
)

print(f"вњ… Dataset: {len(dataset):,} examples formatted.")
print(f"   Sample (first 250 chars):\n   {dataset[0]['text'][:250]!r}")

# в”Ђв”Ђ 8. Optimizer guard в”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђ
# Confirm 8-bit optimizer is reachable before committing to a 3-epoch run.
# paged_adamw_32bit is the safe fallback вЂ” slower but always works.

try:
    _ = bnb.optim.Adam8bit
    OPTIM = "adamw_8bit"
    print("\nвњ… Optimizer: adamw_8bit")
except Exception:
    OPTIM = "paged_adamw_32bit"
    print("\nвљ пёЏ  Optimizer: falling back to paged_adamw_32bit")

# в”Ђв”Ђ 9. SFTConfig в”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђ
# TRL 0.18вЂ“0.21 parameter placement:
#   max_seq_length     в†’ stays on SFTTrainer
#   dataset_text_field в†’ stays on SFTTrainer
# (In TRL >=0.22 these migrated into SFTConfig; with 0.21 pinned, keep them
#  on SFTTrainer to avoid a silent no-op or TypeError.)

sft_config = SFTConfig(
    # в”Ђв”Ђ Output в”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђ
    output_dir                  = OUTPUT_DIR,

    # в”Ђв”Ђ Batch / accumulation в”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђ
    per_device_train_batch_size = BATCH_SIZE,
    gradient_accumulation_steps = GRAD_ACC_STEPS,

    # в”Ђв”Ђ Learning rate в”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђ
    learning_rate               = LEARNING_RATE,
    num_train_epochs            = NUM_EPOCHS,
    warmup_steps                = WARMUP_STEPS,
    lr_scheduler_type           = "cosine",

    # в”Ђв”Ђ Precision: FP16 only вЂ” BF16 explicitly off for T4 в”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђ
    fp16                        = True,
    bf16                        = False,

    # в”Ђв”Ђ Optimizer в”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђ
    optim                       = OPTIM,

    # в”Ђв”Ђ Checkpointing & logging в”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђ
    save_strategy               = "epoch",
    logging_steps               = LOG_STEPS,
    report_to                   = "none",

    # в”Ђв”Ђ Stability flags в”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђ
    # packing=True в†’ variable-length batch tensors в†’ shape mismatches on T4.
    # Keep False until the rest of the pipeline is proven stable.
    packing                     = False,

    # group_by_length: batches similar-length sequences together.
    # Reduces padding waste without the risks of packing.
    group_by_length             = True,

    # Explicit column cleanup prevents the collator from tripping on
    # unexpected fields if the dataset schema changes.
    remove_unused_columns       = True,

    # dataloader_num_workers=0: single-process data loading.
    # Colab's multiprocessing (fork) is unreliable with HuggingFace datasets;
    # worker counts >0 cause occasional silent hangs or CUDA context errors.
    dataloader_num_workers      = 0,

    # pin_memory=False: pinned memory + Colab's CUDA allocator = rare but
    # hard-to-reproduce RuntimeErrors. The throughput gain is negligible
    # for this workload size.
    dataloader_pin_memory       = False,

    # Reproducibility
    seed                        = 42,
    data_seed                   = 42,
)

# в”Ђв”Ђ 10. SFTTrainer в”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђ

trainer = SFTTrainer(
    model              = model,
    train_dataset      = dataset,
    tokenizer          = tokenizer,
    args               = sft_config,
    # в†“ These belong on SFTTrainer (not SFTConfig) for TRL 0.18вЂ“0.21
    dataset_text_field = "text",
    max_seq_length     = MAX_SEQ_LENGTH,
)

# в”Ђв”Ђ 11. Pre-flight report в”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђ

allocated_gb = torch.cuda.memory_allocated() / 1e9
print(f"\nв”Ђв”Ђ Pre-Training Memory в”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђ")
print(f"   GPU total     : {gpu_mem_gb:.1f} GB")
print(f"   Allocated now : {allocated_gb:.2f} GB")
print(f"   Headroom      : {gpu_mem_gb - allocated_gb:.2f} GB")
print(f"   Effective batch: {BATCH_SIZE} Г— {GRAD_ACC_STEPS} = "
      f"{BATCH_SIZE * GRAD_ACC_STEPS}")
print(f"в”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђ\n")

# в”Ђв”Ђ 12. Train в”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђ

print("рџљЂ Training started...\n")
trainer_stats = trainer.train()

# в”Ђв”Ђ 13. Post-training summary в”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђ

peak_mem_gb   = torch.cuda.max_memory_reserved() / 1e9
train_min     = round(trainer_stats.metrics.get("train_runtime", 0) / 60, 2)
train_loss    = round(trainer_stats.metrics.get("train_loss", float("nan")), 4)

print(f"\nв”Ђв”Ђ Training Complete в”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђ")
print(f"   Runtime      : {train_min} min")
print(f"   Final loss   : {train_loss}")
print(f"   Peak VRAM    : {peak_mem_gb:.2f} GB / {gpu_mem_gb:.1f} GB")
print(f"в”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђ\n")

# в”Ђв”Ђ 14. Save в”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђ

print(f"рџ’ѕ Saving adapter to '{ADAPTER_DIR}'...")
model.save_pretrained(ADAPTER_DIR)
tokenizer.save_pretrained(ADAPTER_DIR)
print(f"вњ… Adapter saved to '{ADAPTER_DIR}'")
```
